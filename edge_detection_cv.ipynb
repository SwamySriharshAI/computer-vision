```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow

def display_edge_detection(image, horizontal_mask, vertical_mask):
    # Apply horizontal Prewitt operator mask
    horizontal_edges = cv2.filter2D(image, -1, horizontal_mask)

    # Apply vertical Prewitt operator mask
    vertical_edges = cv2.filter2D(image, -1, vertical_mask)

    # Display the images side by side
    fig, axes = plt.subplots(1, 3, figsize=(10, 5))

    axes[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    axes[0].set_title('Original Image')

    axes[1].imshow(horizontal_edges, cmap='gray')
    axes[1].set_title('Horizontal Edges')

    axes[2].imshow(vertical_edges, cmap='gray')
    axes[2].set_title('Vertical Edges')

    for ax in axes:
        ax.axis('off')

    plt.tight_layout()
    plt.show()
```

```python
# Image Reference: https://www.pxfuel.com/en/desktop-wallpaper-tpjza
image = cv2.imread('/content/charminar.jpeg', cv2.IMREAD_GRAYSCALE)
```

```python
# Prewitt Operator Masks
horizontal_mask = np.array([[1, 1, 1], [0, 0, 0], [-1, -1, -1]])
vertical_mask = np.array([[1, 0, -1], [1, 0, -1], [1, 0, -1]])

display_edge_detection(image, horizontal_mask, vertical_mask)
```

```python
# Sobel Filters
horizontal_mask = np.array([[1, 2, 1], [0, 0, 0], [-1, -2, -1]])
vertical_mask = np.array([[1, 0, -1], [2, 0, -2], [1, 0, -1]])

display_edge_detection(image, horizontal_mask, vertical_mask)
```

```python
# Scharr Filters
horizontal_mask = np.array([[3, 10, 3], [0, 0, 0], [-3, -10, -3]])
vertical_mask = np.array([[3, 0, -3], [10, 0, -10], [3, 0, -3]])

display_edge_detection(image, horizontal_mask, vertical_mask)
```
